```python
#convering the date from DD/MM/YYYY to YYYY-MM-DD
date_info=pd.to_datetime(air_quality_data['Date'])
print(date_info)
```

```python
time_info=air_quality_data['Time']
```

```python
print(time_info)
```

```python
#lambda function for standard
time_info=time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

pandas has dataframe and series

```python
#combining two series to a pandas dataframe:date and time
date_time=pd.concat([date_info,time_info],axis=1)
```

```python
date_time.shape
```

```python
#combining date and time
date_time['ds']=date_time['Date'].astype(str)+" "+date_time['Time'].astype(str)
```

we have to convert ds from object date time to date time format

```python
data=pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y']=air_quality_data['RH']
```

```python
data.head()
```

```python
pip install prophet
```

```python
from prophet import Prophet
```

```python
#training the prophet model
model=Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast=model.predict()
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```
